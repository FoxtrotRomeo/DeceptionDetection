```python
import pandas as pd
```

```python
questionnaire = pd.read_csv('full_dataset.csv')
# create a subset of the questionnaire dataframe containing only the columns 'Participant ID', 'Dyad Number', 'Truth/Lie', 'appreciation_4'
questionnaire_subset = questionnaire[['Participant ID', 'Dyad Number', 'Truth/Lie', 'appreciation_3', 'appreciation_4']]
```

```python
# create a function that can check if the participant which "Participant ID" ends in 'A' for every dyad has an appreciation4 score that agrees with the truth value

def check_appreciation4(dataframe):
    accord_lie = 0
    accord_truth = 0
    disaccord_lie = 0
    disaccord_truth = 0

    total_truth = 0
    total_lie = 0
    # iterate over the rows in the dataframe
    for index, row in dataframe.iterrows():
        # transform the row into a dictionary
        row = dict(row)
        # get the participant ID
        participant_id = row['Participant ID']
        # get the dyad ID
        dyad_id = row['Dyad Number']
        # get the appreciation4 score
        appreciation4 = row['appreciation_4']
        if appreciation4 < 4:
            bin_appr4 = 'Lie'
        elif appreciation4 > 4:
            bin_appr4 = 'Truth'
        elif appreciation4 == 4:
            bin_appr4 = 'Neutral'
        # get the truth value
        truth_value = row['Truth/Lie']
        # check if the participant ID ends in 'A'
        if participant_id[-1] == 'A':
            # check if the appreciation4 score agrees with the truth value
            if bin_appr4 == 'Lie':
                if truth_value == 'Lie':
                    accord_lie += 1
                else:
                    disaccord_lie += 1
            if bin_appr4 == 'Truth':
                if truth_value == 'Truth':
                    accord_truth += 1
                else:
                    disaccord_truth += 1
        if truth_value == 'Truth':
            total_truth += 1
        if truth_value == 'Lie':
            total_lie += 1

    print('Total Truth:', total_truth)
    print('Total Lie:', total_lie)
    return accord_lie, accord_truth, disaccord_lie, disaccord_truth
```

```python
print(check_appreciation4(questionnaire_subset))
```

```python
# create a function that can check if the participant which "Participant ID" ends in 'A' for every dyad has an appreciation4 score that agrees with the truth value

def check_appreciation3(dataframe):
    detect_lie = 0
    detect_truth = 0
    miss_lie = 0
    miss_truth = 0

    total_truth = 0
    total_lie = 0
    # iterate over the rows in the dataframe
    for index, row in dataframe.iterrows():
        # transform the row into a dictionary
        row = dict(row)
        # get the participant ID
        participant_id = row['Participant ID']
        # get the dyad ID
        dyad_id = row['Dyad Number']
        # get the appreciation4 score
        appreciation3 = row['appreciation_3']
        if appreciation3 < 4:
            bin_appr3 = 'Lie'
        elif appreciation3 > 4:
            bin_appr3 = 'Truth'
        elif appreciation3 == 4:
            bin_appr3 = 'Neutral'
        # get the truth value
        truth_value = row['Truth/Lie']

        if truth_value == 'Truth':
            total_truth += 1
        if truth_value == 'Lie':
            total_lie += 1
        # check if the participant ID ends in 'A'
        if participant_id[-1] == 'B':
            # check if the appreciation4 score agrees with the truth value
            if truth_value == 'Lie':
                if bin_appr3 == 'Lie':
                    detect_lie += 1
                else:
                    miss_lie += 1
            elif truth_value == 'Truth':
                if bin_appr3 == 'Truth':
                    detect_truth += 1
                else:
                    miss_truth += 1
    print('Total Truth:', total_truth)
    print('Total Lie:', total_lie)
    return detect_lie, detect_truth, miss_lie, miss_truth
```

```python
true_neg, true_pos, false_pos, false_neg = check_appreciation3(questionnaire_subset)
precision_pos = true_pos / (true_pos + false_pos)
precision_neg = true_neg / (true_neg + false_neg)

recall_pos = true_pos / (true_pos + false_neg)
recall_neg = true_neg / (true_neg + false_pos)

accuracy = (true_pos + true_neg) / (true_pos + true_neg + false_pos + false_neg)

print('Precision Positive:', precision_pos)
print('Precision Negative:', precision_neg)
print('Recall Positive:', recall_pos)
print('Recall Negative:', recall_neg)
print('Accuracy:', accuracy)
```
