```python
import numpy as np
import pandas as pd
import sktime
import os
import glob
```

```python
# read all the csv files in openface_out as dataframes
path = 'opensmile_out'
files = glob.glob('opensmile_out/*.csv')

# create a dictionary called dfs to store all the dataframes
dfs = {}

# read all the csv files in the path
for file in files:
    # use the filename as the key for the dictionary
    key = os.path.basename(file)
    # read the csv file and store it in the dictionary
    dfs[key] = pd.read_csv(file)

    
```

```python
# if the key is longer than 9 characters, append the dataframe to the one with the same first 5 characters, and then the string .csv.
#the dataframe with the longer key has to be at the end of the dataframe with the shorter key.
# the original dataframe is then removed from the dictionary.
for key in list(dfs.keys()):
    if len(key) > 9:
        dfs[key[:5]+'.csv'] = pd.concat([dfs[key[:5]+'.csv'], dfs[key]], axis=0, ignore_index=True)
        del dfs[key]
```

```python
# modify all the keys to remove the .csv extension
for key in list(dfs.keys()):
    dfs[key[:-4]] = dfs.pop(key)

#print the keys and the dimensions of the dataframes
for key in list(dfs.keys()):
    print(key)
    print(dfs[key].shape)
```

```python
# Create a copy of dfs called dfs_to_cut
dfs_to_cut = dfs.copy()
```

```python
# select the lowest number of rows in the dataframes
min_rows = min([dfs[key].shape[0] for key in list(dfs.keys())])

print(min_rows)
```

```python
# delete the rows in the dataframes that are more than the minimum number of rows
for key in list(dfs.keys()):
    dfs[key] = dfs[key].iloc[:min_rows]
    # print the dimensions of the dataframes
    print(dfs[key].shape)
    # reset the index of the dataframes
    dfs[key] = dfs[key].reset_index(drop=True)
```

```python
# check the number of nan values in the dataframes
for key in list(dfs.keys()):
    print(key)
    print(dfs[key].isna().sum().sum())
```

```python
# print the names of the columns in the first dataframe, one per row.
for col in dfs[list(dfs.keys())[0]].columns:
    print(col)
```

```python
# create a variable containing the first dataframe
df = dfs[list(dfs.keys())[0]]
```

```python
# open the full_dataset.csv file as a dataframe, keeping only the first 4 columns
full_df = pd.read_csv('full_dataset.csv', usecols=range(4))

# change the value 'I264B' in the column 'Participant ID' to 'I256B'
full_df['Participant ID'] = full_df['Participant ID'].replace('I264B', 'I256B')
```

```python
#full_df contains a column called "Dyad Number". Use the numbers in this column as keys for a dictionary called dyads.
dyads = {key: [] for key in full_df['Dyad Number'].unique()}
```

```python
# get a list of all the keys in the dfs dictionary.
keys = list(dfs.keys())

# create a dictionary called data_A containing the dataframes where the key ends with "A".
# create a dictionary called data_B containing the dataframes where the key ends with "B".
data_A = {key: dfs[key] for key in keys if key.endswith('A')}
data_B = {key: dfs[key] for key in keys if key.endswith('B')}

# for each dyad number in the full_df dataframe, get the dataframes which names correspond to that number in full_df['Participant ID'] from data_A and data_B.
# If the full_df['Participant ID'] ends with "A", append the dataframe from data_A to the list in the dyads dictionary.
# If the full_df['Participant ID'] ends with "B", append the dataframe from data_B to the list in the dyads dictionary.
for index, row in full_df.iterrows():
    if row['Participant ID'].endswith('A'):
        dyads[row['Dyad Number']].append(data_A[row['Participant ID']])
    elif row['Participant ID'].endswith('B'):
        dyads[row['Dyad Number']].append(data_B[row['Participant ID']])

    
```

```python
for key in dyads.keys():
    print(key)
    # print the dimensions of the dataframes in the list in the dyads dictionary
    for df in dyads[key]:
        print(df.shape)
        # print the number of rows with nan values in the dataframe
        print(df.isna().sum().sum())
```

```python
# save the first dataframe of each listin the folder openface_out_A, using the key as the filename. if the folder does not exist, create it.
save_path = 'opensmile_out_A/'
if not os.path.exists(save_path):
    os.makedirs(save_path)

for key in dyads.keys():
    # convert key to string
    keystr = str(key)
    # save the dataframe, using save_path as the path and keystr as the filename

    dyads[key][0].to_csv(save_path + keystr + '.csv')
```

```python
# Merge the two dataframes in each list in the dyads dictionary, using the timestamp column as the key.
for key in dyads:
    # merge the two dataframes
    dyads[key] = dyads[key][0].join(dyads[key][1], lsuffix='_A', rsuffix='_B')

df = dyads[5]
```

```python
# print the number of nan values in each column of the dataframe in the dyads dictionary.
for key in dyads:
    print(key)
    print(dyads[key].isna().sum())
```

### Let's go back to the dfs_to_cut dictionary

```python
# For each df in the dictionary, let's delete the first 100*60=6000 rows (the first minute) and the last 100*60=6000 rows (the last minute).
for key in dfs_to_cut.keys():
    dfs_to_cut[key] = dfs_to_cut[key].iloc[6000:-6000]
```

```python
# Create a new dictionary called dfs_cut, which contains for each key in dfs_to_cut a list of dataframes, each containing 16000 rows.
dfs_cut = {key: [] for key in dfs_to_cut.keys()}
for key in dfs_to_cut.keys():
    for i in range(0, len(dfs_to_cut[key]), 16000):
        dfs_cut[key].append(dfs_to_cut[key].iloc[i:i+16000])
```

```python
# for each key in dfs_cut, check the length of each dataframe in the list. if the length is not 16000, drop the dataframe.
# if the length is 16000, keep the dataframe and reset the index.
for key in dfs_cut.keys():
    for i in range(len(dfs_cut[key])):
        if len(dfs_cut[key][i]) != 16000:
            dfs_cut[key].pop(i)
        else:
            dfs_cut[key][i] = dfs_cut[key][i].reset_index(drop=True)
```

```python
#full_df contains a column called "Dyad Number". Use the numbers in this column as keys for a dictionary called dyads.
dyads_cut = {key: [] for key in full_df['Dyad Number'].unique()}
```

```python
# get a list of all the keys in the dfs dictionary.
keys = list(dfs_cut.keys())

# create a dictionary called data_A containing the dataframes where the key ends with "A".
# create a dictionary called data_B containing the dataframes where the key ends with "B".
data_A_cut = {key: dfs_cut[key] for key in keys if key.endswith('A')}
data_B_cut = {key: dfs_cut[key] for key in keys if key.endswith('B')}

# for each dyad number in the full_df dataframe, get the dataframes which names correspond to that number in full_df['Participant ID'] from data_A and data_B.
# If the full_df['Participant ID'] ends with "A", append the dataframe from data_A to the list in the dyads dictionary.
# If the full_df['Participant ID'] ends with "B", append the dataframe from data_B to the list in the dyads dictionary.
for index, row in full_df.iterrows():
    if row['Participant ID'].endswith('A'):
        dyads_cut[row['Dyad Number']].append(data_A_cut[row['Participant ID']])
    elif row['Participant ID'].endswith('B'):
        dyads_cut[row['Dyad Number']].append(data_B_cut[row['Participant ID']])
```

```python
# save the first dataframe of each listin the folder openface_out_A, using the key as the filename. if the folder does not exist, create it.
save_path = 'opensmile_out_A_cut/'
if not os.path.exists(save_path):
    os.makedirs(save_path)

for key in dyads_cut.keys():
    # convert key to string
    keystr = str(key)
    for i in range(len(dyads_cut[key][0])):
        number_of_df = '_' + str(i)
        # save the dataframe, using save_path as the path and keystr as the filename

        dyads_cut[key][0][i].to_csv(save_path + keystr + number_of_df + '.csv')
```

```python
# Merge the two dataframes in each list in the dyads dictionary, using the timestamp column as the key.
for key in dyads_cut.keys():
    if len(dyads_cut[key][0]) == len(dyads_cut[key][1]):
        # merge the dataframes in each list
        for i in range(len(dyads_cut[key][0])):
            dyads_cut[key][0][i] = dyads_cut[key][0][i].join(dyads_cut[key][1][i], lsuffix='_A', rsuffix='_B')
    else:
        print('The two lists do not have the same length')
```
